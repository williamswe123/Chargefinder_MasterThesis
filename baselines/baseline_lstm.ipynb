```python

import sys
import os
import torch.optim as optim
import torch.nn as nn
import torch
import matplotlib.pyplot as plt
import numpy as np

from torch.optim.lr_scheduler import ReduceLROnPlateau
from torch.optim.lr_scheduler import LambdaLR

# Get the directory containing the notebook
notebook_dir = os.path.dirname(os.path.abspath("__file__"))

# Add the directory containing the notebook to sys.path
sys.path.append(notebook_dir)

# Add the parent directory (which contains the 'dataloaders' directory) to sys.path
parent_dir = os.path.abspath(os.path.join(notebook_dir, '.'))
sys.path.append(parent_dir)

from lstm import do_da_test

import time
```

```python
seeds = [42, 43, 44]

results = []
for seed in seeds:
    start_time = time.time()  # Start time
    best_model, train_losses, val_losses, best_epoch = do_da_test("varnamo/data_IONITY_LOCF.csv", epochs=15, verbose=False, seed=seed)

    end_time = time.time()
    elapsed_time = end_time - start_time
    
    
    print(f"\tMSE for seed {seed}: {val_losses[best_epoch]}")
    results.append(val_losses[best_epoch])
    
avg = sum(results) / len(results)
std_div = np.std(results)

print("average:" , avg, ", std div:", std_div)
print()
```
