```python
import matplotlib.pyplot as plt
import numpy as np

# Data without 'Scratch'
x = ['1%', '5%', '20%', '50%']
models = ['Finetune-all', 'Finetune-Last-Transformer', 'Finetune-Last-GCN', 'MLP', 'Tiny-Attention', 'BAM']
y_data = {
    '1%': [58.37, 62.26, 95.89, 71.53, 63.07, 63.14],
    '5%': [54.45, 57.26, 68.25, 58.03, 58.24, 53.7],
    '20%': [50.65, 53.17, 57.56, 52.44, 52.49, 51.22],
    '50%': [50.21, 53.49, 57.56, 51.65, 51.43, 50.09]
}
# Standard deviation data
std_data = {
    '1%': [5.27, 10.54, 49.74, 10.70, 5.39, 7.63],
    '5%': [2.21, 4.53, 16.17, 3.88, 3.22, 1.81],
    '20%': [1.26, 2.8, 5.5, 2.01, 1.59, 1.39],
    '50%': [1.21, 2.74, 6.16, 1.53, 2.78, 1.31]
}

# Define colors for each model
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6']

# Plotting
fig, ax = plt.subplots(figsize=(14, 8))

bar_width = 0.12
x_pos = np.arange(len(x))

for i, (model, color) in enumerate(zip(models, colors)):
    mse_values = [y_data[percent][i] for percent in x]
    std_values = [std_data[percent][i] for percent in x]
    ax.bar(x_pos + i * bar_width, mse_values, width=bar_width, label=model, color=color, edgecolor='black', yerr=std_values, capsize=5)

# Add labels and title
ax.set_xlabel('Data Percentage', fontsize=14, fontweight='bold')
ax.set_ylabel('MSE', fontsize=14, fontweight='bold')
ax.set_title('Värnamo to Varberg', fontsize=16, fontweight='bold')
ax.set_xticks(x_pos + bar_width * (len(models) - 1) / 2)
ax.set_xticklabels(x)
ax.legend(title='Model', fontsize=12, title_fontsize='14')
ax.grid(True, linestyle='--', alpha=0.7)
ax.set_facecolor('whitesmoke')

plt.tight_layout()
plt.show()
```
